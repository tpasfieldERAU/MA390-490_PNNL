```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.signal import find_peaks

from matplotlib import rc
sns.set_palette('colorblind')
sns.set_context('paper', font_scale=1.5)
sns.set_style('white')
rc('font', **{'family': 'serif', 'serif': ['Computer Modern']})
rc('text', usetex=True)
```

```python
bg = np.load("../matrices/isolated-materials/float32/Background.npy")
```

```python
# bgc : background counts
# bgb : background bins
bgc, bgb = np.histogram(bg.reshape(-1,1), bins=1000, range=(9000, 14000))
```

```python
plt.stairs(bgc[1:],bgb[1:],fill=True)
plt.xlim(8000, 15000)
plt.xlabel("Intensity")
plt.ylabel("Counts")
plt.title("Background Intensity Distribution")
plt.show()
```

```python
peaks, _  = find_peaks(bgc, height=10000, distance=150)
plt.stairs(bgc[1:], bgb[1:],fill=True)
plt.plot(bgb[peaks], bgc[peaks], 'x')
plt.show()
print(bgb[peaks])

bg_peaks = bgb[peaks]
```
