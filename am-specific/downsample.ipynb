```python
import numpy as np
import scipy.ndimage as ndi
import matplotlib.pyplot as plt
import cv2 as cv
```

```python
mat = np.load("../matrices/NORMALIZED-RP-STACK.npy")
mat = mat[40:1045, 40:1045, 500]
```

```python
plt.imshow(mat[:,:])
plt.show()
```

```python
dmat = ndi.zoom(mat, 0.02, mode='nearest')
# 0.015385
```

```python
plt.imshow(dmat[:,:])
plt.axis(False)
plt.show()
```

```python
out_img = cv.cvtColor(dmat[:,:], cv.COLOR_GRAY2BGR)
img = cv.cvtColor(mat[:,:], cv.COLOR_GRAY2BGR)
cv.imwrite("./ds_test.png", out_img)
cv.imwrite("./test.png", img)
```
