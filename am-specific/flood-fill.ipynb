```python
import numpy as np  # Needed for importing matrices
import scipy.ndimage as ndi  # Possibly needed for image work, not sure
import matplotlib.pyplot as plt  # Needed for plotting
import cv2 as cv  # Needed for image processing
```

```python
mat = np.load("../matrices/NORMALIZED-RP-STACK.npy")
test_img = mat[:,:,200]
```

```python
plt.imshow(test_img)
plt.show()
```

```python
th_img = (test_img < 150) * test_img
th_img = (th_img > 100)
plt.imshow(th_img*255, cmap='gray')
plt.show()
```

```python
th_img = np.array(th_img, dtype=np.uint8)
th_img = cv.cvtColor(th_img, cv.COLOR_GRAY2BGR)
```

```python
dilate = cv.dilate(th_img, (5,5))
plt.imshow(dilate, cmap='gray')
plt.show()
```

```python
fill = cv.floodFill(dilate, dilate, (400,400), (255,255,255))
```
